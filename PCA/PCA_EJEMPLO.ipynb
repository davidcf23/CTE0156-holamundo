```python
import sklearn #pip install -U scikit-learn
import mglearn #pip install mglearn
from sklearn.datasets import load_breast_cancer #carga el data set 
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
%matplotlib inline 
```

```python
mglearn.plots.plot_pca_illustration() #ejemplo ilustrado de PCA
```

```python
cancer=load_breast_cancer() #carga el data set en cancer
cancer.feature_names.shape #muestra tamaño del vector
```

```python
pca=PCA(n_components=2) #Se inicializa PCA con 2 componentes
pca.fit(cancer.data) #los datos se ajustan al PCA
```

```python
transformada=pca.transform(cancer.data) #Transformamos la data a 2 componentes
```

```python
print(cancer.data.shape) #Comparacion entre los 2 datasets
print(transformada.shape)
```

```python
mglearn.discrete_scatter(transformada[:,0],transformada[:,1], cancer.target) #Se grafica el dataset transformado
plt.legend(cancer.target_names,loc='best')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
```

```python
from sklearn.preprocessing import MinMaxScaler 
escala=MinMaxScaler() #objeto de minmaxscaler
escala.fit(cancer.data) #se ajusta el dataset a la escala
escalada=escala.transform(cancer.data)
pca.fit(escalada)
transformada=pca.transform(escalada)
mglearn.discrete_scatter(transformada[:,0],transformada[:,1], cancer.target)
plt.legend(cancer.target_names,loc='best')
plt.gca()
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
```
